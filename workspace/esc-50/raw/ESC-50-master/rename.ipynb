```python
import os

# change this to your target directory
root_dir = r"C:\Users\Louis\PycharmProjects\HTS-Audio-Transformer\workspace\esc-50\raw\ESC-50-master\Audio_full"

for dirpath, dirnames, filenames in os.walk(root_dir):
    for fname in filenames:
        if ")" in fname:
            old_path = os.path.join(dirpath, fname)
            new_fname = fname.replace(")", "_")
            new_path = os.path.join(dirpath, new_fname)
            print(f"Renaming:\n  {old_path}\n→ {new_path}")
            os.rename(old_path, new_path)
```
